# Add Error Bars to a Line Chart

### Adding error bars can show either the standard deviation, standard error, confidence intervals, or minimum and maximum values of a dataset. 

```python
%matplotlib inline
```

```python
# Add Dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Next, we'll calculate the standard deviation of the y_axis values using the statistics module.
# Get the standard deviation of the values in the y-axis.
stdev = round(statistics.stdev(y_axis),2)
stdev

mean = round(statistics.mean(y_axis),2)
mean
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Using object oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

# Add Error bars to a bar chart

```python
plt.bar(x_axis, y_axis, yerr = stdev, capsize=3)
```

```python
# Create error bar chart using object oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
```
